# Phase 2 : Jointure Spatiale et Agrégation par Territoire

Ce notebook valide le rattachement des points de population aux cantons et le calcul des indicateurs agrégés.

```python
import sys
sys.path.append('..')
import pandas as pd
import geopandas as gpd
from src.aggregation import load_cantons_zio, spatial_join_population_cantons, estimate_school_population, aggregate_by_canton
from pathlib import Path

OUTPUT_DIR = Path("../data/output")
```

## 1. Chargement des données de distance (Phase 1)

```python
df_pop = pd.read_parquet(OUTPUT_DIR / "population_distances.parquet")
cantons = load_cantons_zio()
print(f"Cantons de Zio chargés : {len(cantons)}")
```

## 2. Jointure Spatiale et Estimation Scolaire

On attribue un canton à chaque point et on applique les ratios démographiques.

```python
df_joined = spatial_join_population_cantons(df_pop, cantons)
df_estimated = estimate_school_population(df_joined)
df_estimated.head()
```

## 3. Agrégation Finale par Canton

Calcul des moyennes pondérées.

```python
etablissements = gpd.read_file("../data/Etablissement_scolaire_Zio.gpkg")
res_cantons = aggregate_by_canton(df_estimated, etablissements, cantons)
res_cantons.sort_values("dist_moy_lycee", ascending=False)
```
