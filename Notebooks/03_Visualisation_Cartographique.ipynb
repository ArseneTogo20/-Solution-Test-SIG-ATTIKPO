# Phase 3 : Visualisation et Cartographie

Ce notebook génère les rendus visuels pour le rapport final.

```python
import sys
sys.path.append('..')
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from src.visualization import create_canton_choropleth, create_comparative_visualization
```

## 1. Visualisation Choroplèthe (Primaire)

Affichage de la répartition de l'accessibilité scolaire.

```python
cantons_data = pd.read_parquet("../data/output/agregation_cantons.parquet")
cantons_geo = gpd.read_file("../data/Togo_Cantons.gpkg")
cantons_geo = cantons_geo[cantons_geo["prefecture_nom"] == "Zio"]

merged = cantons_geo.merge(cantons_data, on="canton_nom")
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
merged.plot(column='dist_moy_primaire', cmap='YlOrRd', legend=True, ax=ax)
ax.set_title("Distance moyenne à l'école primaire (mètres)")
ax.set_axis_off()
plt.show()
```

## 2. Analyse Comparative

Comparaison des distances entre les cycles Primaire, Collège et Lycée.

```python
df_sorted = cantons_data.sort_values("dist_moy_primaire")
df_sorted.plot(x='canton_nom', y=['dist_moy_primaire', 'dist_moy_college', 'dist_moy_lycee'], kind='bar', figsize=(12,6))
plt.title("Inégalités d'accès selon le cycle d'enseignement")
plt.ylabel("Distance (m)")
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
